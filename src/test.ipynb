```python
import pandas as pd
import re
import seaborn as sns 
import matplotlib.pyplot as plt

import language_tool_python
from collections import Counter
import nltk
from nltk import bigrams

nltk.download('punkt')

sns.set_style('darkgrid')
sns.set_context('notebook') 
```

```python
with open("model_training_model26.txt", "r") as file:
    file_content = file.read()

data = {
    'epoch': [],
    'batch': [],
    'learning_rate': [],
    'ms_per_batch': [],
    'loss': [],
    'perplexity': []
}

pattern = r'\|\s*epoch\s+(\d+)\s+\|\s+(\d+)/\s*(\d+)\s*batches\s+\|\s+lr\s+(\d+\.\d+)\s+\|\s+ms/batch\s+(\d+\.\d+)\s+\|\s+loss\s+(\d+\.\d+)\s+\|\s+ppl\s+(\d+\.\d+)'

for line in file_content.split('\n'):
    match = re.search(pattern, line)
    if match:
        data['epoch'].append(int(match.group(1)))
        data['batch'].append(int(match.group(2)))
        data['learning_rate'].append(float(match.group(4)))
        data['ms_per_batch'].append(float(match.group(5)))
        data['loss'].append(float(match.group(6)))
        data['perplexity'].append(float(match.group(7)))

df = pd.DataFrame(data)
```

```python
plt.figure()
sns.lineplot(data=df, x="batch", y="loss", hue="epoch")
plt.xlabel("Number of batch")
plt.ylabel("Loss")
plt.title("Loss per epoch")

plt.legend(title='Epoch', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
plt.show()
```

```python
plt.figure()
sns.lineplot(data=df, x="batch", y="perplexity", hue="epoch")
plt.xlabel("Number of batch")
plt.ylabel("Perplexity")
plt.title("Perplexity per epoch")
plt.legend(title='Epoch', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
plt.show()
```

```python
plt.figure()
sns.lineplot(data=df.loc[df["epoch"].isin((1, 2, 3))], x="batch", y="ms_per_batch", hue="epoch")
plt.xlabel("Number of batch")
plt.ylabel("MS")
plt.title("MS in batch per epoch")
plt.legend(title='Epoch', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
plt.show()
```

```python
plt.figure()
sns.lineplot(data=df, x="epoch", y="learning_rate")
plt.xlabel("Number of epoch")
plt.ylabel("Learning Rate")
plt.title("Learning Rate per epoch")
```

```python
data = {
    'epoch': [],
    'loss': [],
    'perplexity': []
}
for i in range(1,11):
    df_epoch = df.loc[df["epoch"] == i]
    df_min_loss = df_epoch["loss"].min()
    df_min_perplexity = df_epoch["perplexity"].min()
    data["epoch"].append(i)
    data["loss"].append(df_min_loss)
    data["perplexity"].append(df_min_perplexity)

df = pd.DataFrame(data)
plt.figure()
sns.lineplot(df, x=df["epoch"], y=df["loss"])
plt.xlabel("Number of epoch")
plt.ylabel("Min Loss")
plt.title("Min Loss per Epoch")
```

```python
plt.figure()
sns.lineplot(df, x=df["epoch"], y=df["perplexity"])
plt.xlabel("Number of epoch")
plt.ylabel("Min Perplexity")
plt.title("Min Perplexity per Epoch")
```

```python
fig, ax1 = plt.subplots()

# Plotting loss on the primary y-axis
ax1.plot(df['epoch'], df['loss'], color='blue', label='Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Min Loss per epoch', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a secondary y-axis for perplexity
ax2 = ax1.twinx()
ax2.plot(df['epoch'], df['perplexity'], color='red', label='Perplexity')
ax2.set_ylabel('Min Perplexity per epoch', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Optional: Adding a title and customizing the layout
plt.title('Best Loss and Perplexity per Epoch')
fig.tight_layout()
```

```python
with open("train.txt") as train_file:
    lines = train_file.read()
```

```python
words = lines.split()
words_counter = Counter(words)
sorted_by_count = words_counter.most_common(20)

items, counts = zip(*sorted_by_count)

fig, ax = plt.subplots()
bars = ax.barh(items, counts)

# Adding the counts on the bars
for bar in bars:
    width = bar.get_width()
    label_x_pos = width if width > 0 else width - 10
    ax.text(label_x_pos, bar.get_y() + bar.get_height() / 2, f'{width}', va='center')

# Inversing the x-axis
# ax.invert_xaxis()
frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)
# Setting labels and title
plt.xlabel('Counts')
plt.ylabel('Words')
plt.title('Top 20 words in Train Corpus')

plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.barh(items, counts)

# Adding value labels
for index, value in enumerate(counts):
    plt.text(value, index, f'{value}')

frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)

plt.xlabel('Counts')
plt.title('Top 20 words in Train Corpus')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest value at the top
plt.show()
```

```python
bigrams = [' '.join(e) for e in nltk.ngrams(lines.split(), 2)]
bigrams_counts = Counter(bigrams)
bigrams_sorted = bigrams_counts.most_common(10)

items, counts = zip(*bigrams_sorted)

plt.figure(figsize=(16, 9))
plt.barh(items, counts)

# Adding value labels
for index, value in enumerate(counts):
    plt.text(value, index, f'{value}')

frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)

plt.xlabel('Counts')
plt.title('Top 10 of 2 consecutive words in Train Corpus')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest value at the top
plt.show()
```

```python
_3grams = [' '.join(e) for e in nltk.ngrams(lines.split(), 3)]
_3grams_counts = Counter(_3grams)
_3grams_sorted = _3grams_counts.most_common(10)

items, counts = zip(*_3grams_sorted)

plt.figure(figsize=(16, 9))
plt.barh(items, counts)

# Adding value labels
for index, value in enumerate(counts):
    plt.text(value, index, f'{value}')

frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)

plt.xlabel('Counts')
plt.title('Top 10 of 3 consecutive words in Train Corpus')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest value at the top
plt.show()
```
